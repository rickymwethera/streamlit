```python

import pandas as pd
import numpy as np

dataframe1 = pd.read_csv('Iris_Data.csv')
dataframe1.head()
```

```python
dataframe1['species'] = dataframe1['species'].map({'Iris-setosa': 1, 'Iris-versicolor': 2, 'Iris-virginica': 3})
```

```python
A = dataframe1.iloc[:, :-1]
b = dataframe1.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(A, b, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1 = RandomForestClassifier()
classifier1.fit(X_train, y_train)

y_pred = classifier1.predict(X_test)
# Accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
print("Prediction Accuracy: ", score)
```

```python
import pickle
pickle_out1 = open("classifier1.pkl", "wb")
pickle.dump(classifier1, pickle_out1)
pickle_out1.close()
```
